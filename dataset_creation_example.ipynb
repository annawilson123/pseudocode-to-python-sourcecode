```python
model = genai.GenerativeModel('gemini-pro')
```

```python
import pandas as pd
```

```python
df = pd.read_csv("drive/MyDrive/dataset.csv")
```

```python
df = df.drop(columns=['instruction', 'input', 'prompt'])
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.iloc[4858:4958]
```

```python
df.shape
```

```python
df.head()
```

```python
algo = []
```

```python
for i in df.index:
  j = str(df['output'][i])
  response = model.generate_content("Write the algorithm for the given Python code:"+j);
  algo.append([i, response.text])
```

```python
algo
```

```python
algorithm = []
```

```python
for [i, j] in algo:
  algorithm.append(str(j))
```

```python
algorithm
```

```python
df.insert(0, "Algorithm", algorithm, True)
```
