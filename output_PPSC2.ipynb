```python
from peft import AutoPeftModelForCausalLM
from transformers import GenerationConfig
from transformers import AutoTokenizer
import torch
tokenizer = AutoTokenizer.from_pretrained("ananyarn/results_packing")

inputs = tokenizer("<s>[INST] Algorithm Function to find the largest subarray within an array of 0s and 1s that contains an equal number of both elements.\n1. Initialize `sum` to 0, which will store the cumulative sum of elements in a subarray.\n2.Initialize `maxsize` to -1, which will track the size of the largest subarray found so far.\n3.Iterate through the array excluding the last element\n4.For each `i` If `arr[i]` is 0, set `sum` to -1, representing a 'negative' contribution to the subarray sum. Otherwise, set `sum` to 1, representing a 'positive' contribution.\n 4. Iterate through the array excluding the last element\n5.For each `j` If `arr[j]` is 0, subtract 1 from `sum`. Otherwise, add 1 to `sum`.\n6.Check if `sum` becomes 0. This signifies a subarray starting from `i` with equal contributions from 0s and 1s.\n7.If `sum` is 0 and the current subarray size (`j-i+1`) is greater than `maxsize`, update `maxsize` and store its starting index in `startindex`.\n8.If no subarray with equal 0s and 1s is found, print 'No such subarray'. Otherwise, print the starting and ending indices of the largest subarray found.\n9.The function returns the size of the largest subarray found (which is stored in `maxsize`).[/INST]", return_tensors="pt").to("cuda")

model = AutoPeftModelForCausalLM.from_pretrained(
    "ananyarn/results_packing",
    low_cpu_mem_usage=True,
    return_dict=True,
    torch_dtype=torch.float16,
    device_map="cuda")

generation_config = GenerationConfig(
    do_sample=True,
    top_k=1,
    temperature=0.1,
    max_new_tokens=25,
    pad_token_id=tokenizer.eos_token_id)

model.eval()
with torch.no_grad():
    print("#"*50)
    print("Fine tuned Model")
    print("#"*50)
    print(tokenizer.decode(model.generate(**inputs, max_new_tokens=512, pad_token_id=2)[0], skip_special_tokens=True))
```
